```python
import cv2
```

```python
cap = cv2.VideoCapture('video.mp4') # 입력할 영상 경로 + 파일명 입력
 
count = 0
 
while(cap.isOpened()):
    ret, image = cap.read()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # grayscale변환
    image = cv2.flip(image, 1)                      # 좌우 반전
    

    
    cv2.imwrite("img_output/frame%d.jpg" %count, image)
    print('Saved frame%d.jpg' %count)
    count += 1
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
        
cap.release()
cv2.destroyAllWindows()
```
